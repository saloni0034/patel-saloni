### Variables Description

  **hotel:** Type of hotel (e.g., Resort Hotel, City Hotel).

  **is_canceled:** Indicates whether the booking was canceled (0 = not canceled, 1 = canceled).

  **lead_time:** Number of days between the booking date and the arrival date.

  **arrival_date_year:** Year of arrival (2015, 2016)

  **arrival_date_month:** Month of arrival (January, February)

  **arrival_date_week_number:** Week no. of the year when the arrival is scheduled.

  **arrival_date_day_of_month:** Day of the month when the arrival is scheduled.

  **stays_in_weekend_nights:** no. of weekend nights (Saturday and Sunday) the guest stayed.

  **stays_in_week_nights:** No. of week nights (Monday to Friday) the guest stayed.

  **adults:** No. of adults in the booking.

  **children:** No. of children in the booking.

  **babies:** No. of babies in the booking.

  **meal:** Type of meal plan selected (e.g: Bed & Breakfast, Half Board).

  **country:** Country of origin of the guest.

  **market_segment:** Market segment through which the booking was made (e.g., Online TA, Offline TA/TO).

  **distribution_channel:** Channel used to distribute the booking (e.g., Direct, Corporate).

  **is_repeated_guest:** it means the guest is repeated guest or not. (0 = no, 1 = yes).

  **previous_cancellations:** No. of previous bookings by the guest that were canceled.

  **previous_bookings_not_canceled:** no. of previous bookings by the guest that was not cancelled.

  **reserved_room_type:** Type of room reserved by the guest.

  **assigned_room_type:** Type of room assign to the guest.

  **booking_changes:** No. of changes made to the booking.

  **deposit_type:** Type of deposit made (e.g: No Deposit, Non Refund).

  **agent:** ID of the travel agent who made the booking.

  **company:** ID of the company who made the booking .

  **days_in_waiting_list:** No. of days booking was in the waiting list before confirmed.

  **customer_type:** Type of customer .

  **adr:** Average Daily Rate – the price per night of the stay.

  **required_car_parking_spaces:** No. of car parking spaces required.

  **total_of_special_requests:** No. of special requests made by the guest.

  **reservation_status**: Current status of reservation (e.g : Check-Out, Canceled).

  **reservation_status_date:** Date when the reservation status was last updated

## ***4. Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables***


#### Chart - 1  Cancellation Analysis

```python
# Chart - 1 visualization code

# prblm1: What element contribute most to booking cancellations? How does lead time impact cancellation rates?

# Calculate cancellation rate
df['cancellation_rate'] = df['is_canceled']

lead_time_cancelation = df.groupby('lead_time')['cancellation_rate'].mean().reset_index()
lead_time_cancelation.head()
```

```python
#visualization
sns.lineplot(x='lead_time', y='cancellation_rate', data=lead_time_cancelation ,color='b')
plt.title('Lead Time vs. Cancellation Rate')
plt.xlabel('Lead Time (days)')
plt.ylabel('Cancellation Rate')
plt.grid(True)
plt.show()
plt.rcParams['figure.figsize']=(10,2)
```

##### 1. Why did you pick the specific chart?

The line chart because it clearly shows the trend between lead time and cancellation rates over a continuous days. we can also make heatmap to visualize positive correlation between high lead time(days) and high cancellation rate but i want to show the trend clearly thats why i chose linechart

##### 2. What is/are the insight(s) found from the chart?

The chart shows us a clear trend where higher lead times(days) are associated with increased cancellation rates. This indicates that customers who book far in advance are more likely to cancel their bookings.

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

Positive : hotels can implement strategies like offering incentives for non-cancelable bookings or flexible rebooking options for early bookings.

Negative : High cancellation rates with longer lead times might result in lost revenue

#### Chart - 2 Seasonality

```python
# Chart - 2 visualization code
# Which months have the highest and lowest booking rates? How do bookings vary by season ?

monthly_bookings = df.groupby('arrival_date_month')['hotel'].count().reset_index() #Group by arrival_date_month to calculate the number of bookings
months_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
monthly_bookings['arrival_date_month'] = pd.Categorical(monthly_bookings['arrival_date_month'], categories=months_order, ordered=True)

# Sort the dataframe by the ordered months
monthly_bookings = monthly_bookings.sort_values('arrival_date_month')

# Display the first few rows to verify
monthly_bookings
```

```python
sns.barplot(x='arrival_date_month', y='hotel', data=monthly_bookings, palette='viridis')
plt.title('Monthly Booking Count')
plt.xlabel('Month')
plt.ylabel('Number of Bookings')
plt.rcParams['figure.figsize']=(10,2)
```

##### 1. Why did you pick the specific chart?

I chose bar chart  to clearly compare the number of bookings across different months.

##### 2. What is/are the insight(s) found from the chart?

Chart represent seasonal trends in bookings, showing peak months are July, August and off months  January, February.

#### Chart - 7 Room Preferences:

```python
#Room Preferences:

# how many total times guests get the different room what they reserved?
df['room_type_assigned_diff'] = (df['reserved_room_type'] != df['assigned_room_type']).astype(int)  #just representing the total values
```

```python
assign_df_count=df['room_type_assigned_diff'].value_counts(0=='No', 1=='Yes') # 0 indicates false means guest got the actual what they reserved and 1 means True they got that many times diffrent room what they reserved?
assign_df_count
print(assign_df_count)


sns.barplot(assign_df_count , palette='viridis')
plt.xticks([0, 1], ['No', 'Yes'], rotation=0)
plt.rcParams['figure.figsize']=(12,6)
plt.title('Total times guests get the different room')
plt.show()




```

```python
#How often do guests get assigned a different room type than what they reserved?
room_type_counts = df.groupby(['reserved_room_type', 'assigned_room_type']).size().unstack().fillna(0) #main chart
room_type_counts
```

```python
room_type_counts.plot(kind='bar', stacked=True, figsize=(12, 7))
plt.title('Reserved vs. Assigned Room Types')
plt.xlabel('Reserved Room Type')
plt.ylabel('Number of Bookings')
plt.legend(title='Assigned Room Type')
plt.show()
```

##### 1. Why did you pick the specific chart?

 A stacked bar chart shows how frequently reserved room types match assigned room types and highlights.
